# Credit Card Kaggle- Fixing Imbalanced Dataset

# Context
It is important that credit card companies are able to recognize fraudlent credit card transactions so that customers are not charged for items that they did not purchase.

# Content
The datastes contains transactions made by credit cards in September 2013 by european customers. This dataset presents transactions that occured in two days. where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced the positive class (frauds) account for 0.172% of all transactions.
It contains only numerical input variables which are the result of PCA transaction. Unfortunately due to confidentiality issues, we cannot provide the original features and more background information about the data.Features V1,V2,.... V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. Feature 'Time' contains the seconds elapsed b/w each transaction and the first tansaction in the dataset. The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost senstive learning. Feature 'Class' is the response variable and it takes value 1 in case of Fraud and 0 otherwise

# Inspiration
Identify fraudlent credit card transactions
Given the class imbalanced ratio, we recommend measuring the accuracy using the Area Under the Precision-Recall Curve(AUPRC). Confusion matrix accuracy is not meaningful for unbalanced classification.

# Another technique

```python
#RandomOverSampler to handle imbalanced data
from imblearn.over_sampling import RandomOverSampler

# Initialize RandomOverSampler
os = RandomOverSampler(sampling_strategy=1)  # or 'auto' for default behavior
#Setting it to 1 means you want to have the same number of samples for both 
#classes (a 1:1 ratio).
# Apply RandomOverSampler to your dataset
X_res, y_res = os.fit_resample(X, Y)

#ratio- Suppose keep ratio as 0.5 for 0 which are initially 100 it will try to
# create 50% of the 1's value. If 1 values is of 500 then 0 will be additionally
# create more 150 that means the count will be 100+150=250
# Suppose we have y value in which 1=500 and 0=100 
#as soon as we apply RandomOverSampler with ratio of 0.5. It will go and see
#the maximum or majority category e.g. 1 is majority category with 500 values
#Then the ratio of 0.5 will go and change 0 values and add more 150 points
#because 0.5 of 500 =250 so 150 more points will be added and now the total
#count is 250
```

```python
X_train_res,y_train_res=os.fit_resample(X,Y)
```

```python
X_train_res.shape,y_train_res.shape
```

```python
# Now we can also try with 0.5
os = RandomOverSampler(sampling_strategy=0.5)
```

```python
X_train1_res,y_train1_res=os.fit_resample(X,Y)
```
